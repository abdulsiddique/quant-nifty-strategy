```python
import pandas as pd
import numpy as np

# Load last valid data
df = pd.read_csv("../DATA/merged/nifty_merged_5min.csv")

print("Initial shape:", df.shape)

# Convert numeric columns safely
num_cols = ["close", "open", "high", "low", "volume", "basis", "PCR"]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# Technical indicators
df["EMA_5"] = df["close"].ewm(span=5, adjust=False).mean()
df["EMA_15"] = df["close"].ewm(span=15, adjust=False).mean()
df["SMA_20"] = df["close"].rolling(window=20).mean()
df["SMA_50"] = df["close"].rolling(window=50).mean()

# ðŸš¨ CRITICAL FIX: only drop rows where CLOSE is missing
df = df.dropna(subset=["close"])

print("After feature engineering:", df.shape)

# Save features
df.to_csv("../DATA/merged/nifty_features.csv", index=False)
```
