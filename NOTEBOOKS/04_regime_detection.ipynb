```python
import pandas as pd

df = pd.read_csv("../DATA/merged/nifty_features.csv")

df["regime"] = "Sideways"
df.loc[df["SMA_20"] > df["SMA_50"], "regime"] = "Bullish"
df.loc[df["SMA_20"] < df["SMA_50"], "regime"] = "Bearish"

df.to_csv("../DATA/merged/nifty_with_regime.csv", index=False)

print("✅ Regime detection done")
```
